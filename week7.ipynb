```python
import matplotlib.pyplot as plt


students = ['John','mathew','dave','manasseh','James','caleb','Lubem','mary','Jay','lucky']
English_scores =[78,90,98,78,69,54,78,91,59,55]
maths_scores = [67,67,86,46,38,98,69,84,90,91]
physics_scores = [78,49,76,87,87,56,54,39,90,81]


plt.figure()
plt.plot(students, physics_scores)
plt.title('physics scores - line chart')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.show()

plt.figure()
plt.bar(students, maths_scores)
plt.title('maths scores - bar chart')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.show()

plt.figure()
plt.pie(English_scores, labels=students,autopct='%1.2f%%')
plt.title('English scores - pie chart')
plt.show()
```
