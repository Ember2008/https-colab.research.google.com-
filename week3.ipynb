```python
#create a list of 5 students
scores = [75,88,92,64,81]
print(scores)

#compute average score
average = sum(scores) / len(scores)
print("average score", average)

#find the highest score
highest = max(scores)
print("highset score:",highest)


#create a dictionary of 5 students scores
students = {
    "james":75,
    "mary":88,
    "john":92,
    "grace":64,
    "daniel":81,
}
print(students)

#find the highest scoring student
highest_student = max(students,key=students.get)
print("highest scoring student:", highest_student)
print("score:", students[highest_student])

#create a tuple of 6 students
subjects = ("mathematics", "english", "physics", "chemistry", "biology")
print(subjects)

#create a set of unique scores
unique_scores = set(students.values())
print (unique_scores)
```
